```python
import pandas as pd
df_dss = pd.read_csv('https://cocl.us/datascience_survey_data', index_col='Unnamed: 0')
df_dss.rename(columns={'Unnamed: 0':'Topic'}, inplace=True)
df_dss.head()
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as pat
```

```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.patches as pat
df_dss = df_dss.sort_values(by='Very interested',axis=0, ascending=False)
df_dss_pr = df_dss/2233
ax = df_dss_pr.plot(kind='bar', 
                figsize=(20,8),
                width=0.8,
                color=['#5cb85c', '#5bc0de', '#d9534f'],
                edgecolor=None,
                )
plt.title('Percentage of the Respondents Interest in Data Science Areas', fontsize=16)
plt.legend(labels=df_dss_pr.columns, loc='upper right', fontsize=14) 
ax.tick_params(labelsize=14)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.axes.get_yaxis().set_visible(False)
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate('{:.2%}'.format(height),(x, y + height + 0.01))
plt.show()
```
